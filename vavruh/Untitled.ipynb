```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
import sympy as sm
X=sm.symbols("x")
def alpha(x):
    a1= 0.277928
    a2=2.30541
    a3=0.386029
    b0=1.46029
    b1=1.41008
    b2=1.13532
    b3=0.191355
    return (a1*x+a2*x**2+a3*x**3)/(b0+b1*x+b2*x**2+b3*x**3)
x=np.linspace(0,20,100)
plt.plot(x,alpha(x),label='α(x)')
plt.legend()
alpha(X)
```

```python
def xi1(x):
    a1=  0.267266
    a2=2.2159
    a3=0.753246
    b0=1.02576
    b1=1.1876
    b2=0.803179
    b3=0.109224
    return (a1*x+a2*x**2+a3*x**3)/(b0+b1*x+b2*x**2+b3*x**3)
plt.plot(x,xi1(x),label='ξ1(x)')
plt.legend()
xi1(X)
```

```python
def epsilon0(x):
    return ((1+x**2)**0.5)-1
plt.plot(x,epsilon0(x),label='$\epsilon_{0}(x)$')
plt.legend()
epsilon0(X)
```

```python
M_M0= 0.2751
R_R0=1.2015 

M_M0/R_R0**2
```

```python
def right(x):
    return M_M0/R_R0**2+x-x
x1=np.linspace(0,2,1000)
```

```python
def y(x):
    return alpha(x)*(epsilon0(x)**2/xi1(x)**2)

eps=1e-4
x0=0
for x in x1:
    if M_M0/R_R0**2-y(x)<eps:
        x0=x
        break
x0
```

```python
plt.plot(x1,y(x1),label="left")
plt.plot(x1,right(x1),label="right")
plt.plot(np.array([x0 for _ in range(len(x1))]),np.linspace(0,M_M0/R_R0**2,len(x1)),label="x0")
plt.legend()
y(X)
#plt.legend()
```

```python
eps=1e-4
x0=0
for x in x1:
    if M_M0/R_R0**2-y(x)<eps:
        x0=x
        break
x0
```

```python
def mu(x0):
    return R_R0*(xi1(x0)/epsilon0(x0))
print(mu(x0))
mu(X)
```
